## Prepare data

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sentence_transformers import SentenceTransformer
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

model = SentenceTransformer("all-MiniLM-L6-v2")

    
def clean_text(text, remove_stopwords=True, apply_lemmatization=True):
        text = text.lower()
        text = re.sub(r'<[^>]+>', ' ', text)
        text = re.sub(r'[^a-zA-Z\s]', ' ', text)
        tokens = nltk.word_tokenize(text)
        
        if remove_stopwords:
            tokens = [w for w in tokens if w not in stop_words]
        if apply_lemmatization:
            tokens = [lemmatizer.lemmatize(w) for w in tokens]
            
        return ' '.join(tokens)

def merge_text(row, cols_name):
    parts = []
    for col in cols_name:
        if pd.notnull(row.get(col, None)):
            parts.append(f"{col.replace('_', ' ').capitalize()}: {row[col]}")
    return ". ".join(parts) + "."


def preprocess_and_encode(df, cols_name, model, remove_stopwords=False, apply_lemmatization=False):
    df = df.copy()
    # Merge các cột văn bản
    df['merged_text'] = df.apply(lambda row: merge_text(row, cols_name), axis=1)
    
    # Làm sạch văn bản
    df['clean_text'] = df['merged_text'].apply(
        lambda x: clean_text(x, remove_stopwords=remove_stopwords, apply_lemmatization=apply_lemmatization)
    )
    
    # Encode bằng mô hình ngữ nghĩa
    embeddings = model.encode(df['clean_text'].tolist(), show_progress_bar=True)
    return embeddings
```

```python
cols_name = df.columns.tolist()
cols_name = [col for col in df.columns if col not in ['course_id', 'course_field']]
# Encode các khóa học đã có field
df_known = df[df['course_field'].notnull()]
known_embeddings = preprocess_and_encode(df_known, cols_name,model)

# Encode các khóa học thiếu field
df_missing = df[df['course_field'].isnull()]
missing_embeddings = preprocess_and_encode(df_missing, cols_name,model)
```

```python
from sklearn.neighbors import NearestNeighbors

nn = NearestNeighbors(n_neighbors=3, metric='cosine')
nn.fit(known_embeddings)

distances, indices = nn.kneighbors(missing_embeddings)

# Gán course_field từ các hàng gần nhất
imputed_fields = []
for idx_list in indices:
    neighbor_fields = [df_known.iloc[i]['course_field'] for i in idx_list]
    merged = list(set(f for sub in neighbor_fields for f in sub))
    imputed_fields.append(merged)

# Cập nhật lại vào dataframe
# Lấy index các dòng bị thiếu course_field
missing_idx = df[df['course_field'].isnull()].index

# Tạo một Series với index giống hệt
imputed_series = pd.Series(imputed_fields, index=missing_idx, name='course_field')

# Gán lại đúng cách bằng loc
df.loc[missing_idx, 'course_field'] = imputed_series
display_df(df)
```

```python
# Danh sách các cột cần xử lý
list_columns = [
    'school_about',
    'school_motto',
    'teacher_name',
    'teacher_about',
    'teacher_job_title',
    'teacher_org_name'
]

# Hàm xử lý danh sách, thay giá trị khuyết bằng "unknown"
def fill_missing_in_list(lst):
    if isinstance(lst, list):
        return ["unknown" if pd.isna(item) else item for item in lst]
    return lst  # Giữ nguyên nếu không phải list

def fill_missing_in_list2(lst):
    if isinstance(lst, list):
        return ["no prerequisite" if pd.isna(item) else item for item in lst]
    return lst  # Giữ nguyên nếu không phải list

# Áp dụng cho từng cột trong danh sách
for col in list_columns:
    df[col] = df[col].apply(fill_missing_in_list)
df['concept_prerequisite'] = df['concept_prerequisite'].apply(fill_missing_in_list2)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def hybrid_impute(df, text_cols, list_cols, id_col, target_col, mapping_df, mapping_left, mapping_right, alpha=0.5, k=5, max_recommend=3):
    """
    Hàm kết hợp content-based và collaborative filtering để điền khuyết dữ liệu.
    """
    df = df.copy()
    df = df.reset_index(drop=True)
    df.loc[:, target_col] = df[target_col].apply(lambda x: x if isinstance(x, list) else [])

    # Xử lý text cho các cột list
    list2str_cols=[]
    for col in list_cols:
        new_col=col+'_str'
        df[new_col] = df[col].apply(lambda x: ' '.join([str(i) for i in x if pd.notna(i)]) if isinstance(x, list) else '')
        list2str_cols.append(new_col)
    
    # Tạo text tổng hợp
    df['combined_text'] = df[text_cols + list2str_cols].fillna('').agg(' '.join, axis=1)

    # Tính TF-IDF
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(df['combined_text'])
    similarity_matrix = cosine_similarity(tfidf_matrix)

    # Collaborative filtering mapping
    item_to_idx = {item_id: idx for idx, item_id in enumerate(df[id_col])}
    idx_to_item = {idx: item_id for item_id, idx in item_to_idx.items()}

    mapping_df = mapping_df.dropna(subset=[mapping_left, mapping_right])
    mapping_df = mapping_df.explode(mapping_right).drop_duplicates()
    item_user_dict = mapping_df.groupby(mapping_left)[mapping_right].apply(set).to_dict()

    # Top phổ biến fallback
    all_targets = mapping_df[target_col].value_counts().index.tolist()

    replaced_indices = []

    def impute_row(row_idx, row):
        if isinstance(row[target_col], list) and len(row[target_col]) > 0:
            return row[target_col]

        replaced_indices.append(row_idx)

        # Lấy top-k item tương tự
        sim_scores = list(enumerate(similarity_matrix[row_idx]))
        sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:k+1]

        cb_recommendations = []
        for idx, score in sim_scores:
            similar_item_id = df.iloc[idx][id_col]
            if similar_item_id in item_user_dict:
                cb_recommendations.extend(list(item_user_dict[similar_item_id]))

        cb_recommendations = list(set(cb_recommendations))

        if len(cb_recommendations) == 0:
            # Fallback: lấy top phổ biến
            cb_recommendations = all_targets[:max_recommend]

        return cb_recommendations[:max_recommend]

    imputed_values = []
    for idx, row in df.iterrows():
        imputed_values.append(impute_row(idx, row))

    df[target_col] = imputed_values
    df = df.drop(columns=['combined_text'])
    df = df.drop(columns=[col for col in df.columns if col.endswith('_str')])
    print(f"Số dòng được thay thế: {len(replaced_indices)}")
    print("Các dòng được thay thế (index và ID):")
    print(df.loc[replaced_indices, [id_col, target_col]])

    return df

# Lần 1: Điền teacher_id
mapping_df_teacher = df[['course_id', 'teacher_id']].explode('teacher_id').dropna().drop_duplicates()

df_result = hybrid_impute(
    df, 
    text_cols=['course_name', 'course_about', 'course_prerequisites'],
    list_cols=['school_name', 'course_field'],
    id_col='course_id',
    target_col='teacher_id',
    mapping_df=mapping_df_teacher,
    mapping_left='course_id',
    mapping_right='teacher_id'
)

# Lần 2: Điền school_id từ df_result đã cập nhật
mapping_df_school = df_result[['course_id', 'school_id']].explode('school_id').dropna().drop_duplicates()

df_result = hybrid_impute(
    df_result, 
    text_cols=['course_name', 'course_about', 'course_prerequisites'],
    list_cols=['teacher_about', 'teacher_org_name', 'course_field'],
    id_col='course_id',
    target_col='school_id',
    mapping_df=mapping_df_school,
    mapping_left='course_id',
    mapping_right='school_id',
    max_recommend=2
)
```

```python
def fill_attributes_by_id(course_df, id_col, attr_cols):
    # Merge
    df = course_df[[id_col]+attr_cols].dropna().explode([id_col]+attr_cols)
    df = df.drop_duplicates()
    rel_df = course_df[['course_id',id_col]].explode(id_col)
    df = pd.merge(rel_df,df,on=id_col,how='left')
    df = df.groupby('course_id').agg(list).reset_index()
    course_df = course_df.drop([id_col]+attr_cols,axis=1)
    course_df = pd.merge(course_df,df,on='course_id',how='left')
    return course_df
# Bổ sung thông tin trường học
def fill(course_df,entities):
    for ent in entities:
        cols = course_df.columns[course_df.columns.str.startswith(ent)].tolist()
        attr_cols = [item for item in cols if not (isinstance(item, str) and item.endswith('_id'))]
        id_col = ent+'_id'
        course_df = fill_attributes_by_id(course_df,id_col,attr_cols)
    return course_df
entities=['school','teacher']
course_df = fill(df_result,entities)
```

```python
import pandas as pd

def is_effectively_null(x):
    if x is None or (isinstance(x, float) and pd.isna(x)):
        return True
    if isinstance(x, list):
        return any(pd.isna(i) for i in x)
    return False

def compute_null_stats(df):
    cols = df.columns
    isnull_rate = df.isnull().mean()

    effectively_null_rate = {}
    for col in cols:
        # Nếu kiểu dữ liệu của cột là list
        non_null_values = df[col].dropna()
        if not non_null_values.empty and isinstance(non_null_values.iloc[0], list):
            effectively_null_rate[col] = df[col].dropna().apply(is_effectively_null).mean()
        else:
            effectively_null_rate[col] = 0.0

    null_df = pd.DataFrame({
        'column': cols,
        'isnull_rate': [round(isnull_rate[col], 6) for col in cols],
        'effectively_null_rate': [round(effectively_null_rate[col], 6) for col in cols]
    })
    return null_df

null_stats = compute_null_stats(course_df)
display(null_stats)
```
