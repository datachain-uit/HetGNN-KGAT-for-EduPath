```python
import numpy as np
import pandas as pd
from libreco.data import split_by_num_chrono, DatasetFeat
def split_data(data):    
    data.rename(columns={"course_id": "item","user_id":"user","enroll_time":"time"}, inplace=True)
    train_data, test_data = split_by_num_chrono(data, test_size=1)
    train_data.rename(columns={"item":"course_id","user":"user_id","time":"enroll_time"}, inplace=True)
    test_data.rename(columns={"item":"course_id","user":"user_id","time":"enroll_time"}, inplace=True)
    return train_data,test_data
```

```python
def process_data(item_data,user_item_data):
    item_data["item"] = item_data.index
    user_item_data = user_item_data.merge(item_data[['course_id','item']], on="course_id", how="left")
    user_item_data = user_item_data.groupby("user_id")["item"].apply(list).reset_index()
    return user_item_data
```

```python
def recommend_courses(model, user_course , num_recommendations=None):
    recommendations = {}

    # Dự đoán cho tất cả người dùng trong test_user với thanh tiến trình
    for user_id in tqdm(user_course['user_id'].unique(), desc="Khuyến nghị cho người dùng", unit="user"):
        # Lấy danh sách khóa học mà người dùng đã thích
        like_list = user_course.loc[user_course['user_id'] == user_id, 'item'].iloc[0]
        like_list = like_list[-2:] if len(like_list) >= 2 else like_list
        # Dự đoán cho tất cả các khóa học của người dùng (tắt chi tiết trong dự đoán)
        index = recommend_model.recommend(likes=like_list, dislikes=[]) # input list of like/dislike index (here we like title[10000])
        # Lấy top k khóa học dựa trên dự đoán
        if num_recommendations is None:
            recommended_courses = [titles[i] for i in index]
        else:
            recommended_courses = [titles[i] for i in index[:num_recommendations]]
        # Lưu các khóa học được khuyến nghị
        recommendations[user_id] = recommended_courses

    return recommendations
```

```python
import numpy as np

# Precision@5
def precision_at_k(y_true, recommendations, k=5):
    precision_scores = []
    for user_id, recommended_courses in recommendations.items():
        true_courses = y_true.get(user_id, [])
        if not true_courses:
            continue
        top_k_pred = recommended_courses[:k]
        relevant = sum(1 for course in top_k_pred if course in true_courses)
        precision_scores.append(relevant / k)
    return np.mean(precision_scores) if precision_scores else 0.0

# Recall@5
def recall_at_k(y_true, recommendations, k=5):
    recall_scores = []
    for user_id, recommended_courses in recommendations.items():
        true_courses = y_true.get(user_id, [])
        if not true_courses:
            continue
        top_k_pred = recommended_courses[:k]
        relevant = sum(1 for course in top_k_pred if course in true_courses)
        recall_scores.append(relevant / len(true_courses))
    return np.mean(recall_scores) if recall_scores else 0.0

# Hit Rate@5
def hit_rate_at_k(y_true, recommendations, k=5):
    hit_rate_scores = []
    for user_id, recommended_courses in recommendations.items():
        true_courses = y_true.get(user_id, [])
        if not true_courses:
            continue
        top_k_pred = recommended_courses[:k]
        relevant = any(course in true_courses for course in top_k_pred)
        hit_rate_scores.append(int(relevant))
    return np.mean(hit_rate_scores) if hit_rate_scores else 0.0

# NDCG@5
def ndcg_at_k(y_true, recommendations, k=5):
    dcg_scores = []
    for user_id, recommended_courses in recommendations.items():
        true_courses = y_true.get(user_id, [])
        if not true_courses:
            continue
        top_k_pred = recommended_courses[:k]
        dcg_score = sum(1 / np.log2(i + 2) for i, course in enumerate(top_k_pred) if course in true_courses)
        idcg_score = sum(1 / np.log2(i + 2) for i in range(min(k, len(true_courses))))
        dcg_scores.append(dcg_score / idcg_score if idcg_score > 0 else 0.0)
    return np.mean(dcg_scores) if dcg_scores else 0.0

# MAP@5
def map_at_k(y_true, recommendations, k=5):
    ap_scores = []
    for user_id, recommended_courses in recommendations.items():
        true_courses = y_true.get(user_id, [])
        if not true_courses:
            continue
        top_k_pred = recommended_courses[:k]
        relevant = [1 if course in true_courses else 0 for course in top_k_pred]
        if sum(relevant) == 0:
            ap_scores.append(0.0)
            continue
        precision_at_i = []
        for i in range(min(k, len(top_k_pred))):
            if relevant[i]:
                precision_at_i.append(sum(relevant[:i+1]) / (i+1))
        ap = sum(precision_at_i) / min(len(true_courses), k)
        ap_scores.append(ap)
    return np.mean(ap_scores) if ap_scores else 0.0

# F1 Score@5
def f1_at_k(y_true, recommendations, k=5):
    f1_scores = []
    for user_id, recommended_courses in recommendations.items():
        true_courses = y_true.get(user_id, [])
        if not true_courses:
            continue
        top_k_pred = recommended_courses[:k]
        relevant = sum(1 for course in top_k_pred if course in true_courses)
        precision = relevant / k
        recall = relevant / len(true_courses)
        if precision + recall > 0:
            f1 = 2 * (precision * recall) / (precision + recall)
        else:
            f1 = 0.0
        f1_scores.append(f1)
    return np.mean(f1_scores) if f1_scores else 0.0

# Hàm đánh giá mô hình
def evaluate_model(y_true, recommendations, k=5):
    """
    Đánh giá mô hình với Precision@5, Recall@5, Hit Rate@5, NDCG@5, MAP@5, F1 Score@5.

    Parameters:
    y_true: DataFrame hoặc dict chứa các khóa học thực tế trong tập TEST.
    recommendations: dict chứa các khóa học khuyến nghị.
    k: Số khóa học top cần xét (default là 5).

    Returns:
    dict: Các độ đo Precision@5, Recall@5, Hit Rate@5, NDCG@5, MAP@5, F1 Score@5.
    """
    # Chuyển y_true thành dict
    if hasattr(y_true, 'set_index'):
        y_true_dict = {}
        for user_id, target in y_true.set_index('user_id')['target'].items():
            y_true_dict[user_id] = [target] if not isinstance(target, (list, set)) else list(target)
        y_true = y_true_dict
    else:
        y_true = {k: [v] if not isinstance(v, (list, set)) else list(v) for k, v in y_true.items()}

    precision = precision_at_k(y_true, recommendations, k)
    recall = recall_at_k(y_true, recommendations, k)
    hit_rate = hit_rate_at_k(y_true, recommendations, k)
    ndcg = ndcg_at_k(y_true, recommendations, k)
    map_k = map_at_k(y_true, recommendations, k)
    f1 = f1_at_k(y_true, recommendations, k)

    return {
        f"Baseline": f"V{i+1}-CBF@{k}",
        f"Precision": round(precision,4),
        f"Recall": round(recall,4),
        f"HitRate": round(hit_rate,4),
        f"NDCG": round(ndcg,4),
        f"MAP": round(map_k,4),
        f"F1-score": round(f1,4)
    }
```

```python
def filter_recommendations_from_df(recommendations, train_user_df, test_user_df, all_items, num_neg=100):
    """
    recommendations: dict {user_id: [item1, item2, ..., itemN]}
    train_user_df: DataFrame with columns ['user_id', 'course_id']
    test_user_df: DataFrame with columns ['user_id', 'target'] — 1 positive item per user
    all_items: list or set of all item ids
    """
    assert 'user_id' in train_user_df.columns and 'course_id' in train_user_df.columns
    assert 'user_id' in test_user_df.columns and 'target' in test_user_df.columns

    train_dict = train_user_df.groupby('user_id')['course_id'].apply(set).to_dict()
    test_dict = test_user_df.groupby('user_id')['target'].first().to_dict()

    final_result = {}

    for user_id, rec_items in recommendations.items():
        pos_item = test_dict.get(user_id)
        if pos_item is None:
            continue

        # Ensure pos_item is a scalar
        if isinstance(pos_item, list):
            pos_item = pos_item[0]

        train_items = train_dict.get(user_id, set())
        test_items = {pos_item}
        unavailable_items = train_items.union(test_items)

        candidate_items = list(set(all_items) - unavailable_items)
        if len(candidate_items) < num_neg:
            continue

        sampled_neg_items = set(random.sample(candidate_items, num_neg))
        allowed_items = sampled_neg_items.union({pos_item})

        filtered_items = [item for item in rec_items if item in allowed_items]

        if pos_item in filtered_items and len(filtered_items) >= (num_neg + 1):
            final_result[user_id] = filtered_items

    return final_result
```
