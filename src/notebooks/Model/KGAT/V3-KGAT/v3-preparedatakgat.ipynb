```python
def filter_n_core_interactions(interactions, N_CORE):
    old_size = -1

    while old_size != interactions.shape[0]:
        old_size = interactions.shape[0]

        # Filter users who enrolled in less than n-core courses
        course_counts_of_user = interactions.groupby('user_id').size().reset_index(name='count')
        valid_users = course_counts_of_user[course_counts_of_user['count'] >= N_CORE]['user_id']
        interactions = interactions[interactions['user_id'].isin(valid_users)]

        # Filter courses which were enrolled by less than n-core users
        user_counts_of_course = interactions.groupby('course_id').size().reset_index(name='count')
        valid_courses = user_counts_of_course[user_counts_of_course['count'] >= N_CORE]['course_id']
        interactions = interactions[interactions['course_id'].isin(valid_courses)]
        
        print(f'Old size: {old_size}; New size: {interactions.shape[0]}')

    print('# of users   : ', interactions['user_id'].nunique())
    print('# of courses : ', interactions['course_id'].nunique())

    return interactions

# Ví dụ sử dụng
N_CORE = 5

filtered_interactions = filter_n_core_interactions(user_course, N_CORE)
# filtered_interactions = user_course.copy()
filtered_interactions
```

```python
import pandas as pd

def filter_invalid_relations_and_entities(triplets, min_entities=5, min_rel=25):
    old_size = -1
    while old_size != triplets.shape[0]:
        old_size = triplets.shape[0]

        # Filter entities
        inter_counts_of_entity = triplets['t'].value_counts().reset_index()
        inter_counts_of_entity.columns = ['t', 'count']
        valid_entities = inter_counts_of_entity[inter_counts_of_entity['count'] >= min_entities]['t']
        triplets = triplets[triplets['t'].isin(valid_entities)]

        # Filter interactions
        inter_counts_of_rel = triplets['r'].value_counts().reset_index()
        inter_counts_of_rel.columns = ['r', 'count']
        valid_rels = inter_counts_of_rel[inter_counts_of_rel['count'] >= min_rel]['r']
        triplets = triplets[triplets['r'].isin(valid_rels)]

        print(f'New size: {triplets.shape[0]}')

    print('================ Valid relations ===============')
    print(triplets['r'].value_counts())

    print('===== # of attribute type in each relation =====')
    rels = set(triplets['r'])

    for rel in rels:
        n_uni_attr = triplets[triplets['r'] == rel]['t'].nunique()
        print(f'+ {rel}: {n_uni_attr}')
    
    return triplets

# Ví dụ sử dụng
filtered_triplets = filter_invalid_relations_and_entities(triplets)
# filtered_triplets = triplets.copy()
display(filtered_triplets.head(),filtered_triplets.describe())
```

```python
item = item_df[['org_id']]
entity_df = filtered_triplets[['t']].drop_duplicates().rename(columns={'t': 'org_id'})
entity_df = pd.concat([item, entity_df], ignore_index=True)
entity_df['remap_id'] = range(entity_df.shape[0])
display(entity_df.head(),entity_df.describe())
entity_df.to_csv('kgat_data/entity_list.txt',index=False)
```

```python
rel_df = filtered_triplets[['r']].drop_duplicates()
rel_df['remap_id'] = range(rel_df.shape[0])
rel_df = rel_df.rename(columns={'r': 'org_id'})
rel_df.to_csv('kgat_data/relation_list.txt',index=False)
rel_df.head()
```

```python
entity_mapping = dict(zip(entity_df['org_id'], entity_df['remap_id']))
rel_mapping = dict(zip(rel_df['org_id'], rel_df['remap_id']))
mapping_triplets = filtered_triplets.copy()
mapping_triplets['h'] = mapping_triplets['h'].map(entity_mapping).astype(int)
mapping_triplets['r'] = mapping_triplets['r'].map(rel_mapping).astype(int)
mapping_triplets['t'] = mapping_triplets['t'].map(entity_mapping).astype(int)
```

```python

from tqdm import tqdm
entity_mapping = dict(zip(entity_df['org_id'], entity_df['remap_id']))
user_mapping = dict(zip(user_df['org_id'], user_df['remap_id']))
# Giả sử n_core_interactions là một DataFrame pandas
filtered_interactions['user_id'] = filtered_interactions['user_id'].map(user_mapping).astype(int)
filtered_interactions['course_id'] = filtered_interactions['course_id'].map(entity_mapping).astype(int)

sorted_interactions = filtered_interactions.sort_values('enroll_time')[['user_id', 'course_id']]\
.groupby('user_id')['course_id'].apply(list).reset_index()\
.rename(columns={'course_id': 'course_order'})\
.sort_values('user_id')
display(sorted_interactions.head())
train_file = open('kgat_data/train.txt', 'w')
val_file = open('kgat_data/val.txt', 'w')
test_file = open('kgat_data/test.txt', 'w')

for _, row in tqdm(sorted_interactions.iterrows(), total=sorted_interactions.shape[0]):
    user,courses = row
    train_courses = ' '.join(map(str, courses[:-1]))
    train_file.write(f'{user} {train_courses}\n')
    # val_file.write(f'{user} {courses[-2]}\n')
    test_file.write(f'{user} {courses[-1]}\n')

train_file.close()
# val_file.close()
test_file.close()
```
